# 関数のグラフ

## 関数のプログラム

```python
# x**2 を計算する関数
def f(x):
    return x**2

# f(2) = 2*2 を計算
f(2)
```

```python
from sympy import init_printing, symbols
init_printing(use_latex="mathjax")

x = symbols('x')
```

```python
f = x**2
f
```

```python
#fにx = 2 を代入
f.subs(x, 2)
```

## 関数のプロット

```python
from sympy.plotting import plot
# Jupyter notebookでmatplotlibのグラフを表示
%matplotlib inline
```

```python
# f = x**2 のプロットを作成
plot(f, (x, -2, 2), legend=True)
```

## いろいろな関数とグラフ
### 多項式関数

```python
plot(x + 1, (x, -2, 2), legend=True)
```

```python
plot(x**3 + 2*x**2 - 19*x, (x, -6, 5), legend=True)
```

### 三角関数

```python
# 三角関数と円周率pi をインポート
from sympy import sin, cos, tan, pi
from sympy.plotting import plot_parametric

# 描画サイズを5x5 の正方形に指定する
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 5, 5
```

```python
# 媒介変数t はtheta の頭文字
t = symbols('t')

# 媒介変数表示を使用した単位円のプロットを作成
plot_parametric(cos(t), sin(t), (t, 0, 2*pi), legend=True)

# 描画サイズをdefault 値に戻す
plt.rcParams['figure.figsize'] = 6, 4
```

```python
# sin(x) とcos(x) のプロット
# show=False として、この行ではグラフを表示さない
p = plot(sin(x), cos(x), (x, -2*pi, 2*pi), legend=True, show=False)

# グラフの色を指定
p[0].line_color = 'b' # 青
p[1].line_color = 'r' # 赤

# グラフの表示
p.show()
```

```python
# cos(x) をx 方向に+pi/2 平行移動
cos(x-pi/2)
```

奇関数

偶関数

```python
# tan(x) のプロット
p = plot(tan(x), (x, 0, 2*pi), legend=True, show=False)

# y 軸の表示範囲を指定
p.ylim = (-10, 10)

p.show()
```

### 指数関数

```python
p = plot(2**x, 1/2**x, (x, -2, 2), legend=True, show=False)

p[0].line_color = 'b'
p[1].line_color = 'r'

p.show()
```

### 対数関数

```python
from sympy import log

# log(変数, 底)
log(4, 2)
```

```python
p = plot(log(x, 2), log(2*x, 2), (x, 0.1, 100),
        legend=True, show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'
p.show()
```

```python
p = plot(2**x, log(x, 2), x, (x, -2, 4), legend=True, show=False)
p.ylim=(-2, 4)

p[0].line_color = 'b' # 青
p[1].line_color = 'r' # 赤
p[2].line_color = 'g' # 緑

p.show()
```
