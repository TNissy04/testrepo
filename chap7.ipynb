# 多変数関数の微分

## 2変数関数

### 2変数関数への値の代入
$$z = x^2 - y^2$$

```python
from sympy import init_printing, symbols
init_printing(use_latex='mathjax')

# 変数と関数の定義
x, y = symbols('x y', real=True)
z = x**2 - y**2
z
```

```python
# x = 1, y = -2 の代入
z.subs([(x, 1), (y, -2)])
```

### 2変数関数のプロット

```python
from sympy.plotting import plot3d

# 3 次元プロット
plot3d(z, (x, -5, 5), (y, -5, 5), xlabel='x', ylabel='y')
```

### 2変数関数の理解

```python
import numpy as np

# 1 次元配列をつくった後、2 次元配列をつくる
x_val = np.arange(-2, 2, 0.1)
y_val = np.arange(-2, 2, 0.1)
x_val, y_val = np.meshgrid(x_val, y_val)

# z = x**2 - y**2
z_val = x_val**2 - y_val**2
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 8))

# 3 次元プロットz = x**2 - y**2 のプロット
ax = fig.add_subplot(1, 1, 1, title='x**2 - y**2',
projection='3d')
ax.plot_surface(x_val, y_val, z_val)

# グラフのx,y,z のラベル
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# プロット範囲
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_zlim(-4, 4)
plt.show()
```

$y=a$のとき$z$は下に凸の$x$の2次関数
$$z = x^2 - a^2$$

```python
# 2 変数関数を1 変数関数としてプロットする
fig = plt.figure(figsize=(8, 8))
# y = a 定数とみなしたときのz = x**2 - a**2 のプロット
ax = fig.add_subplot(1, 1, 1, title='x**2 - a**2',
projection='3d')
for i in range(len(x_val)):
    ax.plot(x_val[i], y_val[i], z_val[i], color='c')
# グラフのx,y,z のラベル
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# プロット範囲
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_zlim(-4, 4)
plt.show()
```

$x=a$のとき$z$は上に凸の$y$の2次関数
$$z = -y^2 + a^2$$


```python
# 2 変数関数を1 変数関数としてプロットする
fig = plt.figure(figsize=(8, 8))
# x = a 定数とみなしたときのz = -y**2 + a**2 のプロット
ax = fig.add_subplot(1, 1, 1, title='-y**2 + a**2',
projection='3d')
for i in range(len(x_val[:,])):
    ax.plot(x_val[:,i], y_val[:,i], z_val[:,i], color='c')
# グラフのx,y,z のラベル
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# プロット範囲
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_zlim(-4, 4)
plt.show()
```

## 全微分

$$df = \frac{\partial f}{\partial x}dx + \frac{\partial f}{\partial y}dy$$

$$z_{2} = -x^{2} + xy - y^{2} $$

```python
z2 = -x**2 + x*y - y**2

# x 方向の偏微分
diff(z2, x)
```

```python
# y 方向の偏微分
diff(z2, y)
```

```python
plot3d(z2, (x, -5, 5), (y, -5, 5), xlabel='x', ylabel='y')
```

```python
def plane_tangent(z, a, b):
    pdz_x = z.diff(x).subs(((x, a), (y, b)))
    pdz_y = z.diff(y).subs(((x, a), (y, b)))
    return pdz_x*(x - a) + pdz_y*(y - b) + z.subs(((x, a), (y, b)))
```

```python
plot3d(z2, plane_tangent(z2, 1, -2), (x, -5, 5), (y, -5, 5),
       xlabel='x', ylabel='y')
```

## 偏微分を使う(1) 極大・極小

### 極大・極小の計算

```python
z3 = x**3 + y**3 - 3*x*y

# 3 次元グラフの作成
plot3d(z3, (x, -2, 2), (y, -2, 2))
```

```python
# 停留点を計算
from sympy import solve

# f_x = f_y = 0 を満たす点を求める
para = solve([z3.diff(x), z3.diff(y)], [x, y])
para
```

```python
# 判別式
def discrim (f, a):
    _d = (f.diff(x).diff(y))**2 - f.diff(x, 2) * f.diff(y, 2)
    return _d.subs([(x, a[0]), (y, a[1])])
```

```python
# (x, y) = (0, 0) は判別式> 0　極大・極小ではない
discrim(z3, para[0])
```

```python
# (x, y) = (1, 1) は判別式< 0　極大・極小をとる
discrim(z3, para[1])
```

```python
z3.diff(x, 2).subs([(x, para[1][0]), (y, para[1][1])])
```

```python
# (x, y) = (1, 1) は　極小値
z3.subs([(x, para[1][0]), (y, para[1][1])])
```

## 偏微分を使う(2) 最小二乗法

### ダミーデータの作成

```python
# ダミーデータを作成
from numpy.random import default_rng

# 乱数生成の設定
rng = default_rng(1)

# ダミーデータサイズ
N_t = 16

# 0 < x < 1の範囲で乱数を生成
x_n = rng.random(N_t)
# データのソート
x_n = np.sort(x_n)

# データ0.3x - 1 にノイズを加える
t_n = 0.3 * x_n - 1 + rng.normal(0, 0.03, N_t)

# ダミーデータのプロット
fig = plt.figure()
plt.scatter(x_n, t_n)
```

### 損失関数の計算

```python
# 損失関数E の計算
a, b = symbols('a b', real=True)

E = 0

for i in range(N_t):
    E = E + (t_n[i] - (a * x_n[i] + b))**2

E = E / N_t
```

```python
# 式計算
from sympy import simplify

simplify(E).evalf(1)
```

```python
# 損失関数 E の3次元プロット
plot3d(E, (a, -7, 2), (b, -10, 10), xlabel='a', ylabel='b')
```

### 損失関数の極小値の計算

```python
# 停留点の計算
para_lsm = solve([E.diff(a), E.diff(b)], (a, b))
para_lsm
```

```python
# 近似直線
y_n = para_lsm[a] * x_n + para_lsm[b]

# ダミーデータのプロット 
plt.scatter(x_n, t_n)

# 近似直線のプロット
plt.plot(x_n, y_n)

plt.show()
```

```python
from sympy import Function, Sum
x = Function('x')
t = Function('t')
a, b = symbols('a b')
n, N = symbols('n N', integer=True)

# Sum は和の計算
E = Sum((t(n) - (a*x(n) + b))**2, (n, 1, N)) / N
E
```

```python
E_a = diff(E, a)
E_b = diff(E, b)
```
